## **4. Analysis and Interpretation**  
The Airline Reservation System reveals several operational, technical, and strategic issues affecting overall efficiency, customer satisfaction, and financial performance. A deeper analysis of the system structure, constraints, and data flow highlights critical gaps that need resolution:  

### **1. Overbooking Issue**  
Overbooking occurs when multiple reservations are confirmed for the same seat due to poor seat assignment control and weak uniqueness constraints. This issue arises when the system does not enforce a `UNIQUE` constraint on seat numbers within the same flight. As a result, multiple passengers may be assigned the same seat, leading to confusion during check-in and increased customer dissatisfaction. Strengthening seat assignment validation and adding conflict resolution logic are necessary to resolve this problem.  

### **2. Data Validation Gaps**  
The presence of duplicate email and phone entries in the `Passengers` table reflects weak data validation at the point of data entry. Without proper `UNIQUE` constraints and validation rules, the system allows the creation of multiple records with the same contact details, leading to incorrect customer communication and confusion. Regex-based validation for email and phone numbers, combined with duplicate-checking mechanisms, can improve customer data accuracy and integrity.  

### **3. Crew Scheduling Issue**  
Flights without assigned pilots or crew members indicate poor crew scheduling and operational oversight. The system does not have a mandatory constraint to ensure that a minimum number of crew members are assigned to each flight. This gap increases the risk of last-minute cancellations and flight delays, affecting the airline's operational reliability. Enforcing crew assignment constraints and implementing automated crew scheduling algorithms would enhance operational control and minimize disruptions.  

### **4. Performance Issues**  
Missing indexes on key fields such as `flight_id`, `passenger_id`, and `reservation_id` lead to slow query response times, particularly when handling large datasets. Without indexing, the system scans the entire table for each query, increasing data retrieval time and reducing system efficiency. Adding indexes on frequently queried fields and optimizing query structures would improve data retrieval speed and system responsiveness.  

### **5. Reservation Conflict**  
Passengers with overlapping flight reservations reflect weak reservation conflict handling in the system. The absence of a conflict-checking mechanism allows the same passenger to book multiple flights scheduled at overlapping times. This creates a scenario where passengers may fail to show up for flights or cause rebooking complications. Introducing a conflict-checking algorithm at the time of booking would prevent such issues and improve seat availability management.  

### **6. Payment Issues**  
Missing foreign key constraints between the `Payments` and `Reservations` tables allow orphaned payment records to exist without linked reservations. This causes mismatches in financial reporting and creates potential reconciliation errors. Additionally, payments not linked to confirmed reservations increase the risk of fraudulent activity. Establishing foreign key constraints and ensuring that payment status is directly linked to reservation confirmation would resolve these inconsistencies.  

### **7. Data Redundancy**  
Poor normalization of data leads to unnecessary duplication of records across multiple tables, increasing storage requirements and reducing data retrieval efficiency. For example, passenger information and flight details are repeated in the `Reservations` table instead of using proper foreign key references. Normalizing these tables and creating lookup references would reduce storage redundancy and improve system performance.  

### **8. Status Update Anomalies**  
Inconsistent status progression, such as a flight being marked as "Landed" followed by "Delayed," reflects logical errors in the status update workflow. The system does not enforce a proper sequence of status updates, allowing invalid transitions. Adding status transition rules and implementing a state-based status update system would ensure logical consistency in flight status reporting.  

### **9. Luggage Violations**  
Frequent violations of luggage weight limits suggest poor enforcement of baggage policies. The system lacks automatic baggage weight validation at the time of check-in, allowing passengers to exceed allowed limits without penalty. Integrating baggage weight checks into the reservation and check-in process, along with automated surcharge calculation for excess baggage, would improve compliance and operational efficiency.  

### **10. Customer Loyalty Gaps**  
Lack of tracking for frequent flyer activity and customer preferences reflects missed opportunities for customer engagement and retention. The system does not store data on repeat customers, preventing the airline from offering targeted incentives or loyalty rewards. Introducing a loyalty program linked to customer profiles and reservation history would improve customer satisfaction and retention rates.  

### **11. Data Security Weakness**  
Sensitive customer data such as email, phone, and passport details are stored without encryption, increasing the risk of data breaches and privacy violations. The system lacks role-based access control and encryption standards, making it vulnerable to cyberattacks. Implementing field-level encryption and access controls would enhance data security and protect customer privacy.  

### **12. Check Constraints Issues**  
Incomplete enforcement of valid payment methods allows invalid data entry in the `Payments` table. For instance, a payment marked as "Confirmed" without a valid payment method creates a reporting discrepancy. Strengthening check constraints to allow only predefined payment methods (e.g., credit card, debit card, UPI) and ensuring status alignment would resolve these issues.  

### **13. Crew Shortage**  
Flights operating without adequate crew members reflect poor crew assignment and scheduling. The system does not enforce a minimum crew-to-flight ratio, leading to last-minute cancellations and flight delays. Introducing a crew scheduling algorithm and linking it to flight status would ensure adequate crew assignment and improve operational consistency.  

### **14. Real-Time Update Issue**  
Lack of real-time data syncing for reservation and payment status updates causes delays in customer notifications and operational decisions. For example, a passenger may receive a confirmation for a canceled flight due to outdated data syncing. Implementing a real-time data sync mechanism between the reservation system and customer interface would eliminate these delays and improve customer satisfaction.  

### **15. Flight Status Conflicts**  
Situations where a departure is recorded after cancellation highlight poor data flow and update logic in the system. The system allows status conflicts due to the absence of conditional update rules. Introducing a state-based update hierarchy, where each status follows a predefined sequence, would prevent such conflicts and improve operational accuracy.

## **Database Structure**  
The system consists of **10 interrelated tables** that cover all major aspects of airline operations:  

1. **Airlines**  
   - Stores airline information such as name, country, and headquarters.  
   - Each airline can have multiple flights.  

2. **Airports**  
   - Contains details of airports, including city, country, and a unique airport code.  
   - Each airport can have multiple departing and arriving flights.  

3. **Flights**  
   - Stores details about flights, including departure and arrival airports, airline, status, and aircraft type.  
   - Connected to the Airlines and Airports tables via foreign keys.  

4. **Passengers**  
   - Stores passenger details like name, contact information, passport number, and nationality.  
   - Each passenger can have multiple reservations.  

5. **Reservations**  
   - Records booking details, including flight, passenger, seat, and class.  
   - Linked to the Passengers and Flights tables.  

6. **Crew**  
   - Stores crew information such as name and role (Pilot, Co-Pilot, Attendant).  
   - Each crew member is linked to a specific flight.  

7. **Payments**  
   - Records payment details including amount, payment method, and payment date.  
   - Linked to the Reservations table.  

8. **Luggage**  
   - Stores details about luggage, including type and weight.  
   - Linked to the Reservations table.  

9. **Flight_Status**  
   - Monitors the real-time status of flights (Delayed, Departed, Landed, Cancelled).  
   - Linked to the Flights table.  

10. **Offers_Discounts**  
   - Tracks offers and discounts available to passengers.  
   - Linked to the Passengers table.
